```python
import pandas as pd
import numpy as np

import utils


```

```python
def normalize_category(df):
    
    cp = df.copy()
    columns = cp.columns
    
    for column in columns:
        if(cp[column].dtypes != 'int64' and cp[column].dtypes != 'float64' and  cp[column].dtypes != 'int32' and  cp[column].dtypes != 'float32'):
            cp = utils.normalization(cp,column)
    
    return cp
```

```python
train = pd.read_csv('../../project/banking_data/loanUnitedTrain.csv', sep=',')
test = pd.read_csv('../../project/banking_data/loanUnitedTest.csv', sep=',')


# train = utils.date_conversion(train,"loan_date")
# test = utils.date_conversion(test,"loan_date")

train = normalize_category(train)
test = normalize_category(test)

print(len(train))
print(len(test))

# train.columns
```

```python
### Split the model

from sklearn.model_selection import train_test_split
# from imblearn.over_sampling import SMOTE

df = train.copy()

# X = df2.drop(columns=['status'])
# y = df2['status']

X = df.drop(columns=['loan_success'])
y = df['loan_success']

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.7,test_size=0.3)

print(len(X_test))
print(len(X_train))
```

```python
### Train the Model

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=15)
model.fit(X_train,y_train)
```

```python
### Scores
score = model.score(X_test,y_test)
y_predicted = model.predict_proba(X_test)
y_predicted_absolute = model.predict(X_test)

print(f"score = {score}")
print(f"prediction = {y_predicted}")
print(f"prediction = {y_predicted_absolute}")
```

```python
### Auc Curve
from sklearn import metrics

y_final = y_predicted.transpose()[0]

len(y_final)

fpr, tpr, _ = metrics.roc_curve(y_test, y_final,pos_label=-1)
roc_auc = metrics.auc(fpr, tpr)

print(roc_auc)

final_df = pd.DataFrame()
final_df['Id'] = X_test["loan_id"]
final_df['Predicted'] = y_final

# final_df


```
